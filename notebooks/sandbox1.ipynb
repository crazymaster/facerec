```python
x = np.array([[1, 2, 3], [4, 5, 6]])
y = np.array([1, 2, 3])
print(np.dot(x, y))
x_norm = (x ** 2).sum(axis=1, keepdims=True) ** .5
y_norm = (y ** 2).sum(axis=0, keepdims=True) ** .5
print(x_norm)
print(y_norm)
np.dot(x, y) / x_norm.T / y_norm
```
